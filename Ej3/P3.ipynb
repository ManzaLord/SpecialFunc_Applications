```python
from sympy import Function, symbols
from scipy.special import sph_harm
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

#Simbolos de las variables
x, theta, phi = symbols('x theta, phi')

#Muestra los indices de legendre
def indicesLegendre(l): 
    for m in range(-l, l + 1):
        P = Function(f'P_{{{l}}}^{{{m}}}')(x)
        display(P)

#Muestra los armonicos
def armonicos(l):
    for m in range(-l, l + 1):
        Y = Function(f'Y_{{{l}}}^{{{m}}}') (theta, phi)
        display(Y)

#Grafica los armonicos
def armonicPlot(l, m):
    #Define las coordenadas independientes en una malla de coordenadas
    theta = np.linspace(0, np.pi, 3000)
    phi = np.linspace(0, 2 * np.pi, 3000)
    theta, phi = np.meshgrid(theta, phi)

    #Define las coordenadas cartesianas en funcion de las coordenadas independientes
    x = np.sin(theta) * np.cos(phi)
    y = np.sin(theta) * np.sin(phi)
    z = np.cos(theta)

    #Crea el armonico esferico y obtiene la parte real
    Y_lm = sph_harm(m, l, phi, theta)
    Y_real = Y_lm.real

    #Normaliza los armonicos
    norm = plt.Normalize(vmin=Y_real.min(), vmax=Y_real.max())
    colors = cm.seismic(norm(Y_real))

    #Grafica la parte real de los armonicos
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, facecolors=colors)
    ax.set_axis_off()
    plt.title(f'Parte real de $Y_{{{l}}}^{{{m}}}(\\theta, \\phi)$')
    plt.show()
```

```python
#Parte 1
indicesLegendre(2)
```
